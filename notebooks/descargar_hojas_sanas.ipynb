```python
import tensorflow_datasets as tfds
import tensorflow as tf
import os

# Carpeta de destino para guardar las imÃ¡genes sanas
output_dir = "../data/processed/healthy"
os.makedirs(output_dir, exist_ok=True)

# Descargar dataset Beans
print("ðŸ“¦ Descargando dataset Beans...")
ds = tfds.load("beans", split="train", as_supervised=True")

# Las etiquetas son:
# 0 = healthy
# 1 = angular_leaf_spot
# 2 = bean_rust

# Filtramos solo las hojas sanas
healthy_ds = ds.filter(lambda img, label: tf.equal(label, 0))

# Guardar las imÃ¡genes en formato JPG
print("ðŸ’¾ Guardando imÃ¡genes sanas...")
for i, (img, _) in enumerate(healthy_ds):
    path = f"{output_dir}/healthy_{i}.jpg"
    tf.keras.utils.save_img(path, img)

print(f"âœ… Listo. Se guardaron {i+1} imÃ¡genes en {output_dir}")
```
